```python
'''
整数 N が与えられるので、2 の N 乗を 1000003 で割った余りを求めてください。
・ 0 ≦ N ≦ 10 ^ 12
------ 疑似コード
k = 1000003
余りをｍとする。
２＾ｎとｋは互いに素であるので、
ｎ＋１のとき、
　　余りは２ｍ（２ｍ＜ｋのとき）
　　余りは２ｍ－ｋ（２ｍ＞ｋのとき）
　　    --> 2m % k
ｎ＝０のときｍ＝１
for ループｎまで
あとはｎとｍの関係により計算する
'''
import time
s_time = time.time()

n = 100000000

k = 1000003
m = 1

for i in range(n):
    m = (2 * m) % k
    
print(m)

e_time = time.time()
print("calc_time = {} sec".format(e_time - s_time))
```
